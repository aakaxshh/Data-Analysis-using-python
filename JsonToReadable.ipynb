```python
#function to load the json dataset
def load_data (fileName):
    with open(fileName, "r") as f:
        data=json.load(f)
    return data
    
```

```python
#function to display users, their friends and liked pages
def display_users(data):
    print("users, their connection and like pages")
    for user in data['users']:
        print(f"ID:{user['id']} {user['name']} is friend with {user['friends']} and likes page {user['liked_pages']}")

    print("\npages information")
    for page in data['pages']:
        print(f"{page['id']} represent {page['name']}")
display_users(data)
```
